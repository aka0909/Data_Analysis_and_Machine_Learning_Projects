```python
df1 = pd.read_csv("C:\Users\Akanksha Verma\Desktop\Men's Data\charting-m-matches.csv", delimiter=',', nrows = None, engine= 'python',quoting=csv.QUOTE_NONE)
```

```python
df2 = df1.filter(['match_id','Pl 1 hand','Pl 2 hand', 'Surface'], axis=1)
```

```python
match_points = pd.read_csv("C:\Users\Akanksha Verma\Desktop\Men's Data\charting-m-points.csv", delimiter=',', nrows = None, engine= 'python',quoting=csv.QUOTE_NONE)
match_points_f = match_points.groupby('match_id')
```

```python
def update_result(a):
    loser = match_points.groupby(match_points['match_id']==a)['SetW'].value_counts().index[-1]
    l = loser[1]
    if l==2.0:
        return 2
    else:
        return 1
        
match_ids_list = df2['match_id'].unique().tolist()

temp = {'match_id': match_ids_list}

winner=pd.DataFrame.from_dict(temp)

#a=[]
#for match in match_ids_list:
    #a.append(update_result(match))
#print(a)
winner['winner']=winner['match_id'].apply(update_result)
winner

   
```

```python
data = pd.merge(df2, winner, on="match_id")
data
```
