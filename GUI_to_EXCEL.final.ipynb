```python
import tkinter
from tkinter import *
import openpyxl
from openpyxl.workbook import Workbook
from openpyxl import load_workbook

#https://www.geeksforgeeks.org/python-simple-registration-form-using-tkinter/
```

```python
wb = load_workbook("C:\Users\Akanksha Verma\Desktop\gui_to_excel.xlsx")
sheet = wb.active
def excel(): 
      
    # resize the width of columns in 
    # excel spreadsheet 
    sheet.column_dimensions['A'].width = 20
    sheet.column_dimensions['B'].width = 20
    sheet.column_dimensions['C'].width = 50
    
  
    # write given data to an excel spreadsheet 
    # at particular location 
    sheet.cell(row=1, column=1).value = "SERVE1"
    sheet.cell(row=1, column=2).value = "SERVE2"
    sheet.cell(row=1, column=3).value = "RALLY"
    
def focus1(event): 
    # set focus on the course_field box 
    serve_1.focus_set() 
  
  
    # Function to set focus 
def focus2(event): 
    # set focus on the sem_field box 
    serve_2.focus_set() 
  
  
    # Function to set focus 
def focus3(event): 
    # set focus on the form_no_field box 
    rally.focus_set() 
        
def clear():
    # clear the content of text entry box 
        serve_1.delete(0, END) 
        serve_2.delete(0, END) 
        rally.delete(0, END) 
def insert(): 
    # if user not fill any entry 
    # then print "empty input" 
    if (serve_1.get() == "" and
        serve_2.get() == "" and
        rally.get() == ""): 
              
        print("empty input") 
  
    else: 
  
        # assigning the max row and max column 
        # value upto which data is written 
        # in an excel sheet to the variable 
        current_row = sheet.max_row 
        current_column = sheet.max_column 
  
        # get method returns current text 
        # as string which we write into 
        # excel spreadsheet at particular location 
        sheet.cell(row=current_row + 1, column=1).value = serve_1.get() 
        sheet.cell(row=current_row + 1, column=2).value = serve_2.get() 
        sheet.cell(row=current_row + 1, column=3).value = rally.get() 
       
  
        # save the file 
        wb.save("C:\Users\Akanksha Verma\Desktop\gui_to_excel.xlsx") 
  
        # set focus on the name_field box 
        serve_1.focus_set() 
  
        # call the clear() function 
        clear() 
      
    
```

```python
if __name__ == "__main__": 
    root = Tk()
    root.title('Tennis_Match_Analysis_Data')
    root.geometry("500x500")
    
    excel() 
    
    root.grid_columnconfigure(0, minsize=20)
    root.grid_columnconfigure(1, minsize=30)
    root.grid_columnconfigure(3, minsize=75)
    root.grid_rowconfigure(0, minsize=50)
    root.grid_rowconfigure(2, minsize=50)
    root.grid_rowconfigure(4, minsize=50)
    root.grid_rowconfigure(6, minsize=50)

    label_1 = Label(root,text = "SERVE1:", font = (None, 15))
    label_1.grid(row=1,column=2,sticky=W)

    serve_1 = Entry(root, width = 30)
    serve_1.grid(row=1, column=4, sticky=W)

    label_2 = Label(root,text = "SERVE2 :", font = (None, 15))
    label_2.grid(row=3,column=2,sticky=W)

    serve_2 = Entry(root,width = 30)
    serve_2.grid(row=3, column=4, sticky=W)

    label_3 = Label(root,text = "RALLY :", font = (None, 15))
    label_3.grid(row=5,column=2,sticky=W)

    rally = Entry(root,width = 30)
    rally.grid(row=5, column=4, sticky=W)
    
    serve_1.bind("<Return>", focus1) 
    serve_2.bind("<Return>", focus2) 
    rally.bind("<Return>", focus3) 
    
    excel()
    
    button = Button(root,text='Enter',command=insert, height = 1, width = 10,font = (None, 15))
    button.grid(row=7, column=3, sticky=W)
    
    root.mainloop()


```
