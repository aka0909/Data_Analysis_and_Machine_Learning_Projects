```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
def preprocess_data():
    print("preprocess_data")
    df = pd.read_excel('C:\Users\Akanksha Verma\Desktop\data_sci_learning_outcome.xlsx', sheet_name='Sheet4')
    df_stu= df.groupby(df['student_id']);
   
    transactions = []
    for stu, stu_df in df_stu:
        list_stu=[]
        for index, row in stu_df.iterrows():
            Concept= str(row['concept_pb']).strip()
            Concept= Concept.encode('utf-8', 'ignore').decode('utf-8')
            Label=keyStr(Concept)
            if Label == "":
                print("no question")
            else:
                list_stu.append(Label)
            list_stu=Remove(list_stu)
        transactions.append(list_stu)
    return transactions

def keyStr(original):
    str1 = original.replace(" ", "")
    str2 = str1.replace(",", "")
    str3 = str2.replace("'", "")
    return str3.strip()

def Remove(duplicate):
    final_list = []
    for num in duplicate:
        if num not in final_list:
            final_list.append(num)
    return final_list
```

```python
def printCsv(transactions):
    download_dir = "C:\Users\Akanksha Verma\Desktop\All_concepts_2.csv" #where you want the file to be downloaded to
    csv = open(download_dir, "w") 
    #"w" indicates that you're writing strings to the file
    for trans in transactions:
        row = ",".join(trans)
        row = row + "\n"
        #print(row)
        csv.write(row)
    return True
```

```python
dataset = pd.read_excel("C:\Users\Akanksha Verma\Desktop\All_concepts_2.xlsx")
transactions = []
total_rows=len(dataset.axes[0])
total_cols=len(dataset.axes[1])
print(total_rows)
print(total_cols)
for i in range(0, total_rows):
    row = []
    for j in range(0,total_cols):
        value = str(dataset.values[i,j])
        if value !="nan":
            row.append(value)
        row=Remove(row)
        transactions.append(row)
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.01, min_confidence = 0.8, min_lift = 8, min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
